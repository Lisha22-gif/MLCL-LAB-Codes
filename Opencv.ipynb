```python
import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('//content//rabbit.jpg')# You can replace this file path by your own file path.
print (img.shape)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # converting BGR to RGB for using matplotlib
plt.axis('off')
plt.show()
```

```python
plt.imshow(img[:,:,1])
plt.axis('off')
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
bearface = img[100:800, 150:900]
plt.imshow(bearface)
plt.axis('off')
plt.show()
```
